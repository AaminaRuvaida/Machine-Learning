# How to deal Imbalance dataset?

```python
import pandas as pd
```

```python
df=pd.read_csv('Diabetes_predict.csv')
df.head()
```

```python
df['Outcome'].value_counts()
```

```python
x=df.drop('Outcome',axis=1)
```

```python
x
```

```python
y=df.Outcome
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Upsampling

```python
from sklearn.utils import resample
```

```python
x=pd.concat([x_train,y_train],axis=1)
```

```python
x
```

```python
not_dia=x[x.Outcome==0]
```

```python
not_dia
```

```python
diabetes=x[x.Outcome==1]
```

```python
diabetes
```

```python
dia_upsampled=resample(diabetes,
                      replace=True,
                      n_samples=len(not_dia),
                      random_state=27)
```

```python
dia_upsampled
```

```python
unsampled=pd.concat([not_dia, dia_upsampled])
```

```python
unsampled
```

```python
unsampled.Outcome.value_counts()  
```

# Downsampling

```python
not_dia_downsampled=resample(not_dia,
                            replace=False,
                            n_samples=len(diabetes),
                            random_state=27)
```

```python
not_dia_downsampled
```

```python
downsampled=pd.concat([not_dia_downsampled, diabetes])
```

```python
downsampled.Outcome.value_counts()
```

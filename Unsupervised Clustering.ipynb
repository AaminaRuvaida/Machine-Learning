# K-Means Clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
df=df[["petal length", "sepal length"]]
df.head(6)
```

```python
model=KMeans(n_clusters=3)
y_model=model.fit_predict(df)
y_model
```

```python
model.labels_
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=model.labels_,cmap='viridis')
```

```python
pd.value_counts(y_model)
```

```python
df[y_model==0]
```

```python
df[y_model==1]
```

# Within Cluster Sum of Square (WCSS)

```python
wcss=[]
K=range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(df)
    wcss.append(km.inertia_)
plt.plot(K,wcss,marker="^",color='green')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow method')
plt.show()
```

# Hierarchical Agglomerative clustering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import load_iris
iris=load_iris()
clustering=AgglomerativeClustering().fit(iris.data)
clustering
```

```python
clustering.labels_
```

```python
plt.scatter(iris.data[:,2],iris.data[:,3],c=clustering.labels_,cmap='viridis')
```

```python
plt.scatter(iris.data[:,0],iris.data[:,1],c=clustering.labels_,cmap='viridis')
```
